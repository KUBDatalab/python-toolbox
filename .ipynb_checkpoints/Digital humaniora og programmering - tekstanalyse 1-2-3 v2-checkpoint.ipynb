### Brug RegEx til at rense din tekst 

[Python-funktionen](https://www.w3schools.com/python/python_functions.asp) 'text_scrubber' neden for bruger vi til at rense teksten for alle andre tegn end bogstaver.

Koden gør brug af python string metoderne [.lower()](https://www.w3schools.com/python/ref_string_lower.asp), [.replace()](https://www.w3schools.com/python/ref_string_replace.asp) og [.join()](https://www.w3schools.com/python/ref_string_join.asp), samt biblioteket 're', Python [REgEx](https://www.w3schools.com/python/python_regex.asp).


RegEx mønsteret er '\b\S+\b', som vi kender fra før.

\b : \b finder positionen ved grænsen af et ord (word boundary). \S: \S matcher ethvert ikke-mellemrum +: + matcher det forrige tegn mellem én og et ubegrænset antal gange, så mange gange som muligt ind til næste tegn. Man siger, at plusset er grådigt. \b : \b finder positionen ved grænsen af et ord (word boundary).

Når du sætter \b\S+\b matcher du fra matcher du alle "ikke mellemumstegn" samt underscore dog ikke symboler som punktummer, kommaer, spørgsmålstegn.

Et andet Regex mønster, der også bliver benyttet til rensning er '\w+'.
\w: \w matcher ethvert alfabetisk bogstav (store og små bogstaver), ethvert tal eller en understregning (_). +: + matcher det foregående tegn en eller flere gange.

Når du sætter \w+ sammen matcher du hele ord sammensat af bogstaver, cifre og understregninger.

I første metode forbliver There-for i et ord

I anden metode bliver det til to ord.

Søg f.eks. efter About two o’clock.

I første metode forbliver o’clock et ord.

I anden metode bliver det til to ord "o clock".

Begge metoder efterlader os med underscores.

```python
import re
def text_scrubber(text_string):
    text_clean = text_string.replace('_','')
    text_clean = ' '.join(re.findall(r'\b(\S+)\b', text_clean))
    return text_clean 
```

Når vi skal udføre rensningen af teksten, skal vi enten bruge et Python [loop](https://www.w3schools.com/python/python_lists_loop.asp) eller [list comprehension](https://www.w3schools.com/python/python_lists_comprehension.asp), fordi rensefunktionen skal påføres hvert enkelt af tektens afsnit, som er gemt i listen, som er gemt i variablen sections. Jeg bruger en list comprehension neden for og inspicerer derefter indholdet af listens første element. 

```python
sections_clean_text = [text_scrubber(i) for i in sections] 
```

# Filtering af stopord
Filtering af [stopord](https://en.wikipedia.org/wiki/Stop_word) udør også meget ofte en del af klargøringsprocessen. Med stopord mener vi små ord, der ikke er betydningsbærende. Det vil f.eks. sige artikler, præpositioner, men en del verber falder også meget ofte ind under kategorien stopord. Substantiver falder ofte udenfor kategorien. Stopordslister bliver benyttet til filtere ord væk, som ikke er interessante eller nødvendige i den analyse, som man foretager, og på den måde eksisterer der ikke én universel stopordsliste. Snarere findes der forskellige erfaringer med, hvordan de anvendes bedst muligt til forskellige formål i forskellige sammenhænge.

Nedenfor indlæser jeg en stopords-txtfil med [.read()](https://www.w3schools.com/python/ref_file_read.asp), splitter teksten til en stopordsliste med [.split()](https://www.w3schools.com/python/ref_string_split.asp) og 
filterer stopordene fra ud af teksterne i de forskellige afsnit ved hjælp af en [funktion](https://www.w3schools.com/python/python_functions.asp) og en [list comprehension](https://www.w3schools.com/python/python_lists_comprehension.asp), førend jeg inspicerer det første tekstafsnit, der nu er uden stopord.

```python
url_stopwords = 'https://sciencedata.dk/shared/5dad7af4b0c257e968111dd0ce19eb99?download'
stopwords = urllib.request.urlopen(url_stopwords).read().decode()
stopwords = stopwords.split()
```

```python
# inspect stopword list
stopwords[0:10]
```

```python
def stopword_filter(text):
    word_list = text.split()
    return ' '.join([i for i in word_list if i.lower() not in stopwords]) 
    
sections_clean_text_wo_stopwords = [stopword_filter(i) for i in sections_clean_text]
sections_clean_text_wo_stopwords[1][0:500]
```

```python
from collections import Counter

# take a slice of the text
text = ' '.join(sections_clean_text_wo_stopwords[1:])

# Split the text into words and apply Counter
word_frequencies = Counter(text.split())

# Calculate relative frequencies
total_words = len(text.split()) # cut the words in the text slice
relative_frequencies = {word: count / total_words for word, count in word_frequencies.items()}
sorted_relative_frequencies = dict(sorted(relative_frequencies.items(), key=lambda item: item[1], reverse=True))

# Print the first 15 words with the highest relative frequencies
top_15 = list(sorted_relative_frequencies.items())[:15]
dict(top_15)
```

# Flere metategn, samt pipes, lister og spørgsmålstegn

I litteratur anvender man ofte sammenligninger til at illustrere pointer tydeligere ved at sætte billeder på det man vil beskrive. Sammenligninger bidrager også til at gøre teksten mere levende og intererssant.

Men regex bliver det en overkommelig opgave at hente eksempler på sammenligninger, fordi vi kan finde tekststrenge som følger mønsteret i en typisk sammenligning.

Vi kan illustrere det på følgende måde. Vi leder efter fraser, hvis mønster enten er as a ... eller as an ....

Der er to forskellige måder.

Første måde er med at bruge pipe |. Pipe betyder "eller". Regexmønsteret vil så se således ud: 'as\sa\s\w+|as\san\s\w+'

Anden måde er at bruge liste tegnet []? Det ser således ud: 'as\sa[n]?\s\w+'.

I listen kan der tilføjes bogstaver, som kan stå på den plads i ordet. Spørgsmålstegnet angiver,at bogstavet måske er der og måske ikke er der.

```python
text = ' '.join(sections_clean_text)
comparison = re.findall(r'as\sa\s\w+', text)
print (comparison[0:10])
```

```python
comparison = re.findall(r'as\sa\s\w+|as\san\s\w+', text)
print (comparison[0:10])
```

```python
comparison = re.findall(r'as\sa[n]?\s\w+', text)
print (comparison[0:10])
```

```python
comparison = re.findall(r'as\s\w+\sas', text)
print (set(comparison))
```

# Firkantede parenteser [A-Z]
Find ord, der begynder med store bogstaver


Mange af disse ord er skrevet med stort, fordi de optræder efter et punktum, og på den måde er de ikke, hvad jeg vil kalde for "ægte" ord med stort.

Hvis man vil bortfiltrere de "uægte" ord fra sin liste, så kan man afsløre dem ved at lave et loop og indsætte en betingelse, der kan tjekke om, ordene skulle være skrevet med småt andre steder i teksterne, fordi hvis de er det, så er de "uægte".

```python
text = ' '.join(sections)
upper_case_word = re.findall(r'[A-Z]\w+', text)


true_upper_case = []
for word in upper_case_word:
    if word.lower() not in text:
        true_upper_case.append(word)
print (true_upper_case[0:10])
```

## Beregn type / token ratio (TTR)

En høj TTR viser, at teksten benytter en række forskellge ord, hvilket indikerer mere komplekst eller forskelligartet indhold. En lav TTR viser, at teksten benytter et begrænset sæt af ord, hvilket indikerer et mere effektivt og gentagende ordbrug.
Formlen er: 


$$
TTR = \frac{\text{Number of Unique Words (Types)}}{\text{Total Number of Words (Tokens)}}
$$

```python
# Total antal af ord
subset['Word_count'] = subset['Clean_text'].apply(lambda x : len(x.split()))
```

```python
# Antallet af unikke ord (Types)
subset['Tokens_count'] = subset['Clean_text'].apply(lambda x : len(set(x.split())))
```

```python
# Beregn ttr
subset['TTR'] = subset['Tokens_count'] / subset['Word_count']
```

```python
# Beregn gennemsnit ttr
average = subset['TTR'].mean()

# Beregn forskellen fra gennemsnittet
subset['TTR_differences'] = subset['TTR'] - average
```

```python
subset['TTR_differences'].plot.bar()
```

# SpaCy - analyser af ordklasser - Part-of-Speech (POS) 
[SpaCy](https://spacy.io/) er et forholdsvist nyt Python bibliotek, der anvendes til Natural Language Processing. Det overtager den plads som NTLK biblioteket har haft gennem mange år. 

Med spaCy får man struktur på alle dele af sin tekst, fordi alle dele bliver mærket op med tags og labels. Strukturen består bl.a. i, at alle ord bliver beskrevet med en ordklasse. Det vil sige, at SpaCy prøver at fortolke om et ord er et verbum, substativ, adjektiv osv., og SpaCy prøver også at tolke og beskrive ordenes indbydedes relation, således at alle [sætningsled](https://www.dsn.dk/wp-content/uploads/2021/01/Saetningsled.pdf) også bliver beskrevet. Læg mærke til at jeg skriver "fortolke" og "tolke", hvilket er, fordi SpaCy fejltolker nogle gange.

I spaCy er der blevet anvendt en international standard / systematik, der hedder [Universal Dependencies](https://universaldependencies.org/) for opmærkningen af [ordklasser](https://universaldependencies.org/u/pos/) og [relationerne](https://universaldependencies.org/docsv1/u/dep/all.html), hvilket åbner for forskellige muligheder, når det kommer til at benytte samme scripts på tekster af forskellige sprog.

Foruden opmærkning af ordklasser (POS tagging) og ordernes relationer (dependency tagging) kan spaCy til en vis grad opmærke enheder (entities), dvs. personer, lokaliteter og organisationer(named entity recognition eller NER).

For at kunne bruge Spacy skal man downloade sprogmodeller, som findes inden for få forskellige sprog. Der er heldigvis relative let tilgængelige forklaringe om, hvordan man kommer i gang med at [downloade og bruge sprogmodellerne](https://spacy.io/models/en), og spaCy har også produceret en god [online tutorial](https://course.spacy.io/en), som kan hjælpe os igang med det tekniske.

Det er ikke altid Spacy's opmærkninger er korrekte og selv om, det er muligt at ændre de labels som er opmærket forkert er det ikke nogen let opgave, og det vil ofte kræve mange interationer og tests førend man kan være helt sikker på at ens tekst er helt rigtigt opmærket.

Her er der et eksempel på, hvordan man kan arbejde med at opmærke labels, som er opmærket forkert. 

```python
import spacy
from spacy.pipeline import EntityRuler

# Load the spaCy model
nlp = spacy.load("en_core_web_sm")

# Define a custom pattern for "Albertus Magnus"
patterns = [{"label": "PERSON", "pattern": "Albertus Magnus"}]

# Create an EntityRuler
ruler = EntityRuler(nlp)
ruler.add_patterns(patterns)

# Register the EntityRuler to the pipeline using a name
nlp.add_pipe("entity_ruler", before="ner")
nlp.get_pipe("entity_ruler").add_patterns(patterns)

# gem teksten fra kapitlet i en variabel
new_text = new_subset['Section'][1]

# Process teksten med den modificerede pipeline
new_doc = nlp(new_text)

# send teksten tilbage til dataframen
new_subset['Section'][1] = new_doc
```

# Bag of words eller "BOWs"
"Bag of words" begrebet, også kaldet "bows", er en metode til at få sin tekst repræsenteret som en række af tal. Formålet med at omdanne tekster til rækker af tal er, at man kan bruge matematiske og statistiske metoder til at mål, veje og sammenligne tekster. Bows-metoden består i at tælle antallet af forskellige ord i en tekst.



Link til data. En csv fil: http://bit.ly/f_stein

Se hvordan en csv fil indlæses her: https://pandas.pydata.org/docs/reference/api/pandas.read_csv.html


```python
import pandas as pd

subset = pd.read_csv(r'C:\Users\lakj\Downloads\out.csv')
```

```python
subset.drop(columns='Unnamed: 0', inplace=True)
```

```python
#subset
```

```python
from sklearn.feature_extraction.text import CountVectorizer

corpus = subset['Clean_text'].to_list() # Klargør til sk-learns CountVectorizer 
vectorizer = CountVectorizer()  
X = vectorizer.fit_transform(corpus) 

feature_names = vectorizer.get_feature_names_out() # Træk features ud 
vectors = X.toarray() # Hent vektorerne

bow_vectors = pd.DataFrame(vectors, columns=feature_names) # Byg en bag of words ( bow )

chapters = subset['Chapter'] # gem  data om kapitler

bows = pd.concat([chapters, bow_vectors], axis=1) # Saml vektorer og kapitel data
```

"bows" indeholder en optælling af, hvor mange gange hvert ord optræder i hvert kapitel.

Kolonnenavnene er lig med ordene og værdierne er lig med, hvor mange gande de optræder i hvert kapitel.

I stedet for at beholde de "rå" optalte tal og arbejde videre med dem, så er det ofte bedre at dividere værdierne (antallet af gange hvert ord optræder ) med antallet af ord i hvert pågældende kapitel. VI kalder det at beregne "ordfrevensen" ( word frequency ). På den måde står talene i relation til hinanden og kan lettere sammenlignes. 

```python
# Beregn word frequency
bows_copy = bows.copy()
bows_word_count = bows_copy.iloc[ : , 1:] # subset data til de værdier, der skal summeres
bows_sum = bows_copy.iloc[ : , 1:].sum(axis=1) # summen af værdierne i hver række ( det vil sige antallet af ord i hvert kapitel)
bow_word_frequency = bows_word_count.div(bows_sum, axis=0) # Beregn word frequency
chapters_bows_word_freq = pd.concat([chapters, bow_word_frequency], axis=1) # Saml kapiter og ordfrekvenser 
```

### Vega-Altair

Vega-Altair er et Python visualiseringsbiblioteket. Det kan f.eks. benyttes til interaktive visualiseringer, som kan være smarte og flotte. I Vega-Altair er der en del ting, der foregår automatisk. Det hænger sammen med, at programmet automatisk fortolker de input, du giver og mapper dem til forskellige visualiseringsmuligheder. Se f.eks. i afsnittet ["Encoding"](https://altair-viz.github.io/user_guide/encodings/index.html) i dokumentationen. Der kan opstå forskellige kringlede situtationer, hvor det fuldautomatiske setup bliver upraktisk. Det sker f.eks., når du arbejder med input, hvor der indgår bogstaver og tal ('chapter 1', 'chapter' 2 osv.) Når Vega-Altair koder inputtet som "Ordinal", bliver inputtet til kategoriske data og bliver sorteret alfabetisk. Derfor kommer 'chapter 10' efter 'chapter 1', da '1' er efterfulgt af '0' i en alfabetisk sortering.

Vi laver en visualisering af, hvor nogle typiske genre-begreber optræder i romanen.

```python
# Tliføj "gotiske" søgeord
terms_to_search = ['dark', 'forest', 'ice', 'lake', 'lightning', 'mist', 'moon', \
                   'night', 'ocean', 'rain', 'river', 'rock', 'sea', 'star', 'storm', \
                   'sunrise', 'sunset', 'thunder', 'water', 'weather', 'winter']

# Transform data til et langt format
long_df = chapters_bows_word_freq.melt(id_vars=['Chapter'], value_vars=terms_to_search, 
                                       var_name='Term', value_name='Frequency')

long_df['Chapter_no'] = long_df['Chapter'].str.extract('(\d{1,2})') # Byg ny kolonnen med kapitel nummerne 
```

```python
# Vega-Altair Line chart
import altair as alt
import warnings
warnings.filterwarnings("ignore")

selection = alt.selection_point(fields=['Term'], bind='legend')

chart = alt.Chart(long_df).mark_line().encode(
    x='Chapter_no:Q', # Vælg Quantitative data til chapter nummerne,
    #x='Chapter:O',  # Vælg Ordinal data for chapters,
    y='Frequency:Q',  # Quantitative data til frequency, og
    color='Term:N',  # Nominal data for terms
    tooltip=['Chapter', 'Term', 'Frequency'],
    opacity=alt.condition(selection, alt.value(0.9), alt.value(0.05)) 
).properties(
    width=600,
    height=300,
    title='Keyword Trend Plot'
).interactive(
).add_params(
    selection
)

chart
```

# Distance-beregning og sammenligning af tekster
BOW og teksternes vektoriseringen giver mulighed for at sammenligne tekster med udgangspunkt i ordernes hyppighed.

Som skrevet så består BOW i en repræsentation af tekst som en række af tal. Når man har to eller flere rækker af tal, så kan man sammenligne talrækkerne parvis og beregne differencen eller afstanden mellem talrækkerne. Til det bruger man euklidiske afstandsberegning. Vi kan applicere afstandsberegningen på vores bow-data ved at importe metoden "euclidean_distances" fra sk-learn.

Resultatet bliver returneret i en tabel eller matriks, som er nærmest umulig at aflæse. Vi kan se resultaterne nedenfor i variablen "similarity_matrix" og "similarity_df". For at komme til en forståelse af afstandene mellem teksterne, så vælger man ofte at visualisere resultatet ved hjælp af enten et heatmap eller et dendrogram, men netværksdiagrammer vil også kunne anvendes fordi, der kan værdien af den parvise relation mellem teksterne illustreres ved hjælp af degrees mellem edges.

Neden for bruger jeg biblioteket Seaborn til et heatmap, bibliotet SciPy til et dendrogram og bibliotekerne NetworkX og NxAltair til et netværksdiagram.

Nu kan vi aflæse af visualiseringerne at kapitel 14 ligger længst væk fra de andre kapitler, mens kapitler som 21 og 23, 16 og 20, samt 18 og 22 ligger tæt på hinanden.

```python
from sklearn.metrics.pairwise import euclidean_distances
import numpy as np

# Compute the pairwise distances between word frequencies
distances = euclidean_distances(chapters_bows_word_freq.iloc[:, 1:])

sigma = 1.0  # Adjust this value to control the spread of the kernel
similarity_matrix = np.exp(-distances ** 2 / (2 * sigma ** 2))

# Convert similarity matrix to a DataFrame
similarity_df = pd.DataFrame(similarity_matrix, index=list(chapters), columns=list(chapters))
```

```python
# The fast
import seaborn as sns
sns.heatmap(similarity_df, annot=False, cmap='coolwarm', square=True, cbar_kws={'shrink': .5})
```

```python
# The slow
import seaborn as sns
import matplotlib.pyplot as plt

# Set up the matplotlib figure
plt.figure(figsize=(6, 5))

# Draw the heatmap
ax = sns.heatmap(similarity_df, annot=False, cmap='coolwarm', square=True, cbar_kws={'shrink': .5})

# Labels and title
plt.title('Chapter Similarity matrix', fontsize=16)
plt.xlabel('Chapters', fontsize=12)
plt.ylabel('Chapters', fontsize=12)

# Adjustments for better visualization
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)

# Show the plot
plt.show()
```

En anden måde at vise afstandene mellem vektorerne (kapitlerne) er at bruge til et dendrogram.

Jeg bruger _method='ward'_, fordi jeg synes metoden ofte giver pænere visualiseringer end andre "linkage" metoder ('single', 'complete', 'average', 'weighted', 'median').

Prøv selv.

```python
import numpy as np
from sklearn.metrics import pairwise_distances
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt

# Create a linkage matrix for hierarchical clustering
linkage_matrix = linkage(similarity_matrix, method='ward')  
# You can choose among the following linkage methods: 'single', 'complete', 'average', 'weighted', 'median', and 'ward'.

# Create and visualize the dendrogram
dendrogram(linkage_matrix, labels=list(chapters), orientation="right")
plt.title('The similarities between the chapters in hierarchical clustering')
plt.show()
```

# Tf-idf

```python
from sklearn.feature_extraction.text import TfidfVectorizer

corpus = list(subset['Clean_text'])

# Initialize the TF-IDF vectorizer - choose a max feature 
tfidf_vectorizer = TfidfVectorizer(max_features=5000, stop_words='english')

# Fit and transform the text data to calculate TF-IDF scores
tfidf_matrix = tfidf_vectorizer.fit_transform(corpus)

# Get the TF-IDF scores for each item in the text_list
tfidf_scores = tfidf_matrix.toarray()

# Initialize a list to store the TF-IDF scores for each item
tfidf_scores_list = []

# Iterate through the text_list and build a list of TF-IDF scores
for i, item in enumerate(corpus):
    words = tfidf_vectorizer.get_feature_names_out()
    scores = tfidf_scores[i]
    # build dictionary
    item_scores = {word: score for word, score in zip(words, scores)}
    # Sort dictionary by value and only append the ten words with the highest value
    item_scores_sorted = sorted(item_scores.items(), key=lambda x:x[1], reverse=True)
    tfidf_scores_list.append(item_scores_sorted[0:10])
    
subset['Distinctive_keywords'] = tfidf_scores_list
```

```python
new_subset = subset[['Chapter', 'Distinctive_keywords']]
```

Kolonnen "Distictive_keywords" indeholder i øjeblikket datalister med tuples. Listerne skal pakkes ud i individuelle rækker, og herefter kan vi oprette et interaktivt diagram med en dropdown-menu, hvor vi kan vælge et kapitel og vise de ord, der kendetegner det.

```python
# Unpacking the Distinctive_keywords column
rows = []
for index, row in new_subset.iterrows():
    for word, score in row['Distinctive_keywords']:
        rows.append({'Chapter': row['Chapter'], 'Keyword': word, 'Score': score})

# Create a new dataframe
keywords_df = pd.DataFrame(rows)
```

```python
# Create a dropdown selection for chapters
chapter_dropdown = alt.binding_select(options=keywords_df['Chapter'].unique(), name='Chapter')
chapter_selection = alt.selection_single(fields=['Chapter'], bind=chapter_dropdown)

# Create a bar chart
chart = alt.Chart(keywords_df).mark_bar().encode(
    x=alt.X('Score:Q', title='TF-IDF Score'),
    y=alt.Y('Keyword:N', title='Distinctive Keywords', sort='-x'),
    color='Chapter:N',
    tooltip=['Keyword', 'Score']
).transform_filter(
    chapter_selection
).properties(
    width=500,
    height=200,
    title="Distinctive Keywords (TF-IDF) by Chapter"
).add_params(
    chapter_selection
)

chart
```
