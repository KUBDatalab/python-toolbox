# Telegram chat export - copy thumbnails

This script finds all files in the `photos` folder where the filename contains "thumb" or "thumbnail" (case-insensitive) and copies them to `thumbnail_photos`. The destination is automatically created if it does not exist.

Thumbnails are smaller files, which therefore take up less space and are often easier to work with.

I would recommend watching this [playlist](https://youtube.com/playlist?list=PLmNPvQr9Tf-aRstY69vGAPO_c5RaceBhN&si=aUMSOPfzmJefq17Q) and find inspiration and ideas for how **Orange can be used** to work further with the collected image material.

```python
# Import libraries
from pathlib import Path
import shutil
from tqdm import tqdm
```

## Source and destination folders

Adjust the paths to the folders to suit your own project.

In the example below, I have edited the first path, 'Al Jazeera English ChatExport_2025-09-02\photos'. I have added 'Al Jazeera English'. The remaining part of the path name comes from the Telegram-Export file that I downloaded from the Telegram channel 'Al Jazeera English', unzipped and placed in a convenient place on my computer.

I have also edited the second path, 'Al Jazeera English ChatExport_2025-09-02\thumbnail_photos'. The first part of the path name points to the same folder as the first path, while the second part of the path name, 'thumbnail_photos', is the name of a folder that is not part of the Telegram-Export file that I downloaded.

The folder 'thumbnail_photos' is created when the script is run.

```python
# Set the source folder containing the photos
source_dir = Path(r'Al Jazeera English ChatExport_2025-09-02\photos')

# Destination folder (will be created inside the source folder)
dest_dir = Path(r'Al Jazeera English ChatExport_2025-09-02\thumbnail_photos')
dest_dir.mkdir(exist_ok=True)

# Get list of files in source_dir
files = list(source_dir.iterdir())

# Initialize counter for copied files
copied = 0

# Iterate with tqdm progress bar
for file_path in tqdm(files, desc="Copying files"):
    if file_path.is_file():
        name_lower = file_path.name.lower()
        if "thumb" in name_lower or "thumbnail" in name_lower:
            shutil.copy2(file_path, dest_dir / file_path.name)
            copied += 1

print(f"Copied {copied} files to: {dest_dir}")
```

You can just run these two cells.
    
But change the paths in `source_dir`and `dest_dir` so it suits your project.
